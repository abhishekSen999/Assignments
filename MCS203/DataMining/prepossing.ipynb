```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#splitting pre normalised data year wise
data=pd.read_csv('data.csv')

states=list(['Andhra Pradesh','Arunachal Pradesh','Assam','Bihar','Chhattisgarh','Goa','Gujarat','Haryana','Himachal Pradesh','Jammu & Kashmir','Jharkhand','Karnataka','Kerala','Madhya Pradesh','Maharashtra','Manipur','Meghalaya','Mizoram','Nagaland','Odisha','Punjab','Rajasthan','Sikkim','Tamil Nadu','Tripura','Uttar Pradesh','Uttarakhand','West Bengal','A&N Islands','Chandigarh','D&N Haveli','Daman & Diu','Delhi UT','Lakshadweep','Puducherry','All India'])

for year in range(2001,2013):
    yy=str(year)
#     yy='2001'
    eachyear=data[['STATE/UT','CRIME HEAD',yy]]
    eachyear=eachyear.set_index('STATE/UT')    
    eachyear=eachyear.set_index('CRIME HEAD').groupby(level=0).apply(lambda x : np.concatenate(x.values.tolist())).apply(pd.Series)
    eachyear.columns=states

    eachyear=eachyear.T
    eachyear=eachyear.drop('All India')
    eachyear=eachyear.drop('TOTAL CRIMES AGAINST WOMEN',axis=1)
    eachyear
    eachyear.to_csv(('DATA'+yy+'.csv'))

    testdata=pd.read_csv(('DATA'+yy+'.csv'))
    testdata=testdata.rename(index=str,columns={'Unnamed: 0':'STATE/UT'})
    testdata.to_csv(('DATA'+yy+'.csv'),index=False)


```

```python
# merging all years
import pandas as pd
import glob

path = 'normalisedallyears' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame=frame.drop('STATE/UT',axis=1)
frame.to_csv('categorised_crime.csv',index=False)
```

## finding PCA on 2001 data

```python
data2001=pd.read_csv("DATA_perlakh_2001.csv")


states=data2001['STATE/UT']
x=data2001.drop('STATE/UT',axis=1)
features=list(x.columns)
x = StandardScaler().fit_transform(x)
pd.DataFrame(data=x,columns=features)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf.insert(loc=0,column='STATE/UT',value=data2001['STATE/UT'],)
principalDf.to_csv("principal_components_2001.csv")
```

# normalizind data for each year

```python
data=pd.read_csv('data.csv')
femalePopulation=pd.read_csv('FemalePopulation.csv')
states=femalePopulation['STATE/UT']

for state in states:
    for year in range(2001,2013):
        data_each_state=data.loc[data['STATE/UT']== state][str(year)]
        population=femalePopulation.loc[femalePopulation['STATE/UT']==state ][str(year)]
        population=population.values[0]
        new_data_each_state=data_each_state/population * 100 # as population is in thousands
        index=new_data_each_state.index.values
        for i in index:
            data.loc[i,str(year)]=new_data_each_state.loc[i]
        
        
data.to_csv('normalised.csv') 
```
